```python
import cv2
import torch
from collections import Counter

# Load the YOLO model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Open the webcam (use `0` for the default webcam, or change for other cameras)
cap = cv2.VideoCapture(0)
```

```python

# Check if the webcam was successfully opened
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Get webcam properties for saving
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Codec for .avi files
output_path = 'output_live.avi'  # Path for saving the output video
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Class categories for counting (YOLO class names are model-specific)
vehicle_classes = ["car", "truck", "bus", "motorbike"]  # Classes to count as vehicles
object_counter = Counter()

# Process video feed frame by frame
while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read frame.")
        break

    # Run YOLO on the current frame
    results = model(frame)

    # Reset the object counter for the current frame
    object_counter.clear()

    # Process detections
    detections = results.xyxy[0]  # Bounding boxes and class info
    for det in detections:
        if len(det) >= 6:  # Ensure valid detection
            x1, y1, x2, y2, conf, cls = det[:6]
            class_name = model.names[int(cls)]
            label = f"{class_name} {conf:.2f}"

            # Update object counter if the detected class is in the vehicle list
            if class_name in vehicle_classes:
                object_counter[class_name] += 1

            # Draw bounding boxes and labels on the frame
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display object counts on the frame
    y_offset = 20  # Initial offset for displaying counts
    for obj, count in object_counter.items():
        count_text = f"{obj.capitalize()}: {count}"
        cv2.putText(frame, count_text, (10, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
        y_offset += 30  # Move the text down for the next object

    # Write the frame with detections to the output video
    out.write(frame)

    # Display the frame with detections
    cv2.imshow("Live Traffic Detection", frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```
