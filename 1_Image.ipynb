# Array Structure
- **Outer Brackets** [[[ ... ]]]: These represent the entire 3D array of pixels in the image.
- **First Layer** [[ ... ], [ ... ], ...]: Each nested list represents a row of pixels.
- **Second Layer** [ ... , ... ]: Within each row, there are further nested lists that represent individual pixels in that row.
- **Third Layer** [B G R]: Inside each pixel list are three values representing the color channels: Blue, Green, and Red intensities, in that order.

e.g. 4x4x3 image

 [
        [[0, 0, 0], [10, 20, 30], [20, 40, 60], [30, 60, 90]],
        [[40, 80, 120], [50, 100, 150], [60, 120, 180], [70, 140, 210]],
        [[80, 160, 240], [90, 180, 255], [100, 200, 255], [110, 220, 255]],
        [[120, 240, 255], [130, 255, 255], [140, 255, 255], [150, 255, 255]]
  ]

  - First layer Entire image [...]
  - second layer [[...],[...],...] i.e [
    [0, 0, 0], [10, 20, 30], [20, 40, 60], [30, 60, 90]] row 1 of 4 rows of image
- Row 1: [[0, 0, 0], [10, 20, 30], [20, 40, 60], [30, 60, 90]]
- Row 2: [[40, 80, 120], [50, 100, 150], [60, 120, 180], [70, 140, 210]]
- Row 3: [[80, 160, 240], [90, 180, 255], [100, 200, 255], [110, 220, 255]]
- Row 4: [[120, 240, 255], [130, 255, 255], [140, 255, 255], [150, 255, 255]]

  - Third layer : Individual pixels [B,G,R]
