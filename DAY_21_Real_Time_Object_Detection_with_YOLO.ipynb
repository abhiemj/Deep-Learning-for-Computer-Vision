```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the YOLOv5 model pretrained on the COCO dataset
model = YOLO('yolov8n.pt')  # YOLOv8 Nano (lightweight for real-time use)
```

```python
# Load an image for detection
image_path = '/content/Transpo-Mumbaitraffic-1079622456.webp'  # Replace with your image file path
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR (OpenCV format) to RGB
plt.imshow(img_rgb)
```

```python
# Perform inference with YOLO
results = model.predict(source=img_rgb, save=False, show=False)
```

```python
# Plot the original image with detection results
annotated_image = results[0].plot()

plt.figure(figsize=(10, 10))
plt.imshow(annotated_image)
plt.title('Object Detection Results')
plt.show()
```
