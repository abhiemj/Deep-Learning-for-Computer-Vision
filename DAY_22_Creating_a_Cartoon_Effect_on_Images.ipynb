```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image
image = cv2.imread('/content/download (1).jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for proper visualization

# Display the original image
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Step 1: Apply bilateral filter to reduce noise and preserve edges
smoothed_image = cv2.bilateralFilter(image, d=9, sigmaColor=75, sigmaSpace=75)

# Step 2: Convert to grayscale and apply edge detection
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
edges = cv2.adaptiveThreshold(
    gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9
)

# Step 3: Combine edges with smoothed image
cartoon_image = cv2.bitwise_and(smoothed_image, smoothed_image, mask=edges)

# Display the cartoon effect
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cartoon_image)
plt.title("Cartoon Effect")
plt.axis("off")
plt.show()
```
